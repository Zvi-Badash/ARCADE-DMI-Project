```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
metrics = pd.read_csv('lightning_logs/version_40/metrics.csv')

# rolling mean impute missing values
metrics = metrics.ffill().bfill()
metrics.describe()

# Plot the loss all columns that have the word loss (after the 50 first steps)
losses = metrics.filter(like='class', axis=1)[:]
losses = losses.rolling(1000).mean()
losses.plot(figsize=(10, 10))
plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('Loss over steps')
plt.show()
```

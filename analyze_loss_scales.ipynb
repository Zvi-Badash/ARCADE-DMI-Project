```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
metrics = pd.read_csv('lightning_logs/version_16/metrics.csv')

# rolling mean impute missing values
metrics = metrics.ffill().bfill()
metrics.describe()

# Plot the loss all columns that have the word loss (after the 50 first steps)
losses = metrics.filter(like='loss', axis=1)[5000:]
# scale columns to have mean 0 and std 1
losses = (losses - losses.mean()) / losses.std()
losses = losses.rolling(1000).mean()
losses.plot(figsize=(10, 5))
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss over epochs')
plt.show()
```
